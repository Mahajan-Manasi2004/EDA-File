# Merge

```python
employee = pd.read_csv('Employee.csv',index_col=0)
salaries = pd.read_csv('Salaries.csv',index_col=0)
```

```python
employee
```

```python
salaries
```

```python
# inner join

pd.merge(employee,salaries,how='inner')
```

```python
# left join

pd.merge(employee,salaries,how='left')
```

```python
# Right join
```

```python
pd.merge(employee,salaries,how='right')
```

```python
# All Outer Join
```

```python
pd.merge(employee,salaries,how='outer')
```
